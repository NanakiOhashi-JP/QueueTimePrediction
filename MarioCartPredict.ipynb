```python
import pandas as pd
import glob

# glob でファイル一覧を取得（必要に応じてパスを修正）
file_pattern = "/content/sample_data/data/download_mariokartkoopaschallenge_*.csv"
csv_files = glob.glob(file_pattern)

# 見つかったファイル一覧を確認
print("Found files:", csv_files)

# もしファイルが見つからなければ、ファイルパスやパターンを再確認
if not csv_files:
    raise FileNotFoundError("CSVファイルが見つかりません。ファイルの存在場所とパターンを確認してください。")

# 各CSVファイルを読み込んでリストに格納
df_list = []
for file in csv_files:
    df = pd.read_csv(file)
    df.columns = df.columns.str.strip()  # カラム名の前後の空白を除去
    df_list.append(df)

# 複数のDataFrameを統合
data = pd.concat(df_list, ignore_index=True)

# 統合後のカラムを確認
print("統合後のカラム:", data.columns)
```

```python
import pandas as pd
import glob

# 既にファイル一覧が取得され、統合後のDataFrame "data" がある前提です。
# （例: 既に df_list から pd.concat() して data ができている）

# Date/Time カラムを datetime 型に変換
# 方法A: フォーマット指定の場合
# data['Date/Time'] = pd.to_datetime(data['Date/Time'], format='%Y-%m-%d %H:%M:%S')

# 方法B: 自動推論の場合（形式にばらつきがあるときはこちらが無難です）
data['Date/Time'] = pd.to_datetime(data['Date/Time'], errors='coerce')

# 変換結果を確認
print("変換後の Date/Time:")
print(data['Date/Time'].head())

# 数値型への変換（念のため）
data['Wait Time'] = pd.to_numeric(data['Wait Time'], errors='coerce')

# 対象のライドでフィルタリング
data = data[data['Ride'] == "Mario Kart: Koopa's Challenge"]

# 日付でソート
data = data.sort_values('Date/Time')

print("フィルタ後の先頭データ:")
print(data.head())
```

```python
# Date/Time をインデックスに設定し、"Wait Time" カラムだけを抽出してリサンプリング
data_resampled = data.set_index('Date/Time')['Wait Time'].resample('15min').mean()

# 必要に応じて、インデックスを DataFrame のカラムに戻す
data_resampled = data_resampled.reset_index()

print(data_resampled.head())
```

```python
# data_resampled は既に Date/Time と Wait Time の2列を持っている前提です。
df_prophet = data_resampled.rename(columns={'Date/Time': 'ds', 'Wait Time': 'y'})

# データを日付順にソート（通常はリサンプリング時に既に整列していますが念のため）
df_prophet = df_prophet.sort_values('ds')

print(df_prophet.head())
```

```python
from prophet import Prophet
import matplotlib.pyplot as plt

# Prophet モデルのインスタンスを作成
model = Prophet()

# モデルの学習
model.fit(df_prophet)

# 未来の期間を定義 (例: 15分刻みで96ポイント＝1日分)
future = model.make_future_dataframe(periods=96, freq='15min')

# 予測を実行
forecast = model.predict(future)

# 予測結果のプロット
fig1 = model.plot(forecast)
plt.title("Mario Kart: Koopa's Challenge Wait Time Forecast")
plt.xlabel("Date")
plt.ylabel("Wait Time")
plt.show()

# コンポーネント（トレンドや季節性など）のプロット
fig2 = model.plot_components(forecast)
plt.show()
```

```python
# 予測結果の CSV 出力
forecast.to_csv("future_forecast.csv", index=False)
```
